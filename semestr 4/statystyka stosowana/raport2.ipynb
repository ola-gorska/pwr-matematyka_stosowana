```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
```

## Zadanie 1

### $\mu \neq 1.5$

```python
df = pd.read_csv('zadanie1.txt', header=None, names=['value'])
data = df['value'].tolist()
```

```python

sample_mean = np.mean(data)
alpha = 0.05
std_dev = 0.2
mu_0 = 1.5
n = len(data)

z = (sample_mean - mu_0)/(std_dev/np.sqrt(n))
left_crit_val = stats.norm.ppf(alpha/2) 
right_crit_val = stats.norm.ppf(1-(alpha/2))

pvalue = 2*(1-stats.norm.cdf(abs(z)))

x_axis = np.linspace(-10,10, 100)
y_axis = stats.norm.pdf(x_axis)

plt.plot(x_axis, y_axis, label=r'Gęstość rozkładu $N(0,1)$')
plt.fill_between(x_axis[x_axis > right_crit_val], y_axis[x_axis > right_crit_val], color='red', label='Obszary krytyczne')
plt.fill_between(x_axis[x_axis < left_crit_val], y_axis[x_axis < left_crit_val], color='red')
plt.scatter(z, 0, marker='^', s=20, color='black')
plt.vlines(z, 0, max(y_axis), color='black', linestyle='--', label='Wartość statystki Z')
plt.title(r'$\alpha = 0.05$')
plt.legend()
```

Wartość statystyki Z: $-7.0415$,

P-value wynosi $1.9025 \cdot 10^{-12}$

```python
df = pd.read_csv('zadanie2.txt', header=None, names=['values'])
data = df['values'].tolist()
```

```python
alpha = 0.05
mu = 0.2
sigma0 = 1.5
n=len(data)

variance = np.var(data, ddof=1) #nieobciążony estymator wariancji

x_squared = (n-1)*variance/sigma0
left_crit_val = stats.chi2.ppf((alpha/2), n-1)
right_crit_val = stats.chi2.ppf(1-(alpha/2), n-1)

pvalue = 2*min(stats.chi2.cdf(x_squared, n-1), 1-stats.chi2.cdf(x_squared, n-1))

x_axis = np.linspace(900, 1150, 100)
y_axis = stats.chi2.pdf(x_axis, n-1)
plt.plot(x_axis, y_axis, label=r'Gęstość rozkładu $\chi^2$')
plt.fill_between(x_axis[x_axis > right_crit_val], y_axis[x_axis > right_crit_val], color='red', label='Obszary krytyczne')
plt.fill_between(x_axis[x_axis < left_crit_val], y_axis[x_axis < left_crit_val], color='red')
plt.scatter(x_squared, 0, marker='^', s=20, color='black')
plt.vlines(x_squared, 0, max(y_axis), color='black', linestyle='--', label='Wartość statystki Z')
plt.title(r'$\alpha = 0.05$')
plt.legend()
```
